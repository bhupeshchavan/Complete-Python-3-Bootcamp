```python
#Global vairable

p1=''
p2=''
p1_mark=''
p2_mark=''
go_first=''
go_sec=''
go_first_turn=0
go_sec_turn=0
turn_p=0
turn_win=0

def tic_toc_game():

    def player_input():
        def p1_input():
            f1=input('Please enter player 1 name:')
            global p1
            p1=f1
            if f1=='':
                print('Name cannot be empty.')
                p1_input()
            #return p1
            
        def p2_input():
            f2=input('Please enter player 2 name:')
            global p2
            p2=f2
            if f2=='':
                print('Name cannot be empty.')
                p2_input()

        def p1_marker():
            f1_mark=input('Please select your marker "X" or "O" :')
            global p1_mark
            p1_mark=f1_mark
            if f1_mark=='':
                print('Marker cannot be empty.')
                p1_marker()
            
        
        def p2_marker():
            f2_mark=input('Please select your marker "X" or "O" :')
            global p2_mark
            p2_mark=f2_mark
            if f2_mark=='':
                print('Marker cannot be empty.')
                p2_marker()
        
        p1_input() 
        p1_marker()
        print(f'{p1}\'s marker is {p1_mark}')
        p2_input()
        p2_marker()
        print(f'{p2}\'s marker is {p2_mark}')


    from IPython.display import clear_output

    def display_board(board):
        print('             ' + '________________')
        print('             ' + '|''   ' +board[10]+'|''   ' +board[10]+'|''   ' +board[10]+' |')
        print('             ' + '|''  ' +board[7]+' |''  '+board[8]+' |''  '+board[9]+'  |')
        print('             ' + '|''   ' +board[10]+'|''   ' +board[10]+'|''   ' +board[10]+' |')
        print('             ' + '|''_______________''|')
        print('             ' + '|''   ' +board[10]+'|''   ' +board[10]+'|''   ' +board[10]+' |')
        print('             ' + '|''  ' +board[4]+' |''  '+board[5]+' |''  '+board[6]+'  |')
        print('             ' + '|''   ' +board[10]+'|''   ' +board[10]+'|''   ' +board[10]+' |')
        print('             ' + '|''_______________''|')
        print('             ' + '|''   ' +board[10]+'|''   ' +board[10]+'|''   ' +board[10]+' |')
        print('             ' + '|''  ' +board[1]+' |''  '+board[2]+' |''  '+board[3]+'  |')
        print('             ' + '|''   ' +board[10]+'|''   ' +board[10]+'|''   ' +board[10]+' |')
        print('             ' + '|''_______________''|')


    test_board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ',' ']

    player_input()
    print ('                 This is a tic toc toe game        ')
    print (f'                its showtime {p1} vs {p2} ')

    #print('Game will start in 10 seconds')

    import random
    import string
    def choose_first(p1,p2):
        a=[p1,p2]
        global go_first
        global go_sec
        global go_first_turn
        global go_sec_turn
        global turn_p
        turn_p=random.choice(a)
        print(turn_p + ' will deliver the first blow')
    

    choose_first(p1,p2)
    display_board(test_board)
    print(f'        {turn_p} will be first.')


    def next_turn(turn):
        global p1
        global p2
        global p1_mark
        global p2_mark
        global turn_p
        global turn_win
        test_board[10]=' '
        turn_p=turn
        position=(input(f'{turn_p} please mark your place:'))
        print(position)
        l_num_chk=['1','2','3','4','5','6','7','8','9']
        if position=='' or position not in l_num_chk or position == '0':
            print('place cannot be null.Please mark your place')
            board_input()
            
        pos=int(position)
        clear_output()
        if test_board[pos]==' ':
            if turn==p1:
                turn_win=turn
                test_board[pos]=p1_mark
                display_board(test_board)
                turn_p=p2
            else:
                turn_win=turn
                test_board[pos]=p2_mark
                display_board(test_board)
                turn_p=p1
        else:
            display_board(test_board)
            print('Uhh..thats already blocked. Choose again')
            board_input()


    def board_input():
        test_board[10]='#'
        global turn_p
        global turn_win
        while ' ' in test_board:
            turn=turn_p
            next_turn(turn)
            test_board[10]='#'
            if test_board[1] == test_board[2] == test_board[3] and test_board[1] != ' ' and test_board[2] != ' ' and test_board[3] != ' ' or \
            test_board[1] == test_board[4] == test_board[7] and test_board[1] != ' ' and test_board[4] != ' ' and test_board[7] != ' ' or \
            test_board[1] == test_board[5] == test_board[9] and test_board[1] != ' ' and test_board[5] != ' ' and test_board[9] != ' ' or \
            test_board[2] == test_board[5] == test_board[8] and test_board[2] != ' ' and test_board[5] != ' ' and test_board[8] != ' ' or \
            test_board[3] == test_board[6] == test_board[9] and test_board[3] != ' ' and test_board[6] != ' ' and test_board[9] != ' ' or \
            test_board[3] == test_board[5] == test_board[7] and test_board[3] != ' ' and test_board[5] != ' ' and test_board[7] != ' ' or \
            test_board[4] == test_board[5] == test_board[6] and test_board[4] != ' ' and test_board[5] != ' ' and test_board[6] != ' ' or \
            test_board[7] == test_board[8] == test_board[9] and test_board[7] != ' ' and test_board[8] != ' ' and test_board[9] != ' ':
                print(f'Hurray..and your winner is {turn_win}')
                again=int(input('Do you want play again:'))
                if again == 0:
                    tic_toc_game()
                else:
                    break

            if ' ' not in test_board:
                clear_output()
                test_board[10]=' '
                display_board(test_board)
                print('Board is full. Its a tie..tough fight')
                again=int(input('Do you want play again:'))
                if again == 0:
                    tic_toc_game()
                else:
                    break
        
        
    board_input()

tic_toc_game()
```
